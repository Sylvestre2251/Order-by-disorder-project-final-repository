# Tutorial for the simulation

Before running the code, one should install plotly with pip install plotly to see in 3D the lattice and spins interactively in browser.
The relevant functions are packaged into Kagome.py.
The simulation class can be initialised just from the lattice number of triangles in the two directions Nx and Ny
and specifying J for the hamiltonian , We also provide a name for the directory in which the files will be created for the results.
- The next two argument allow you to control the seed for the generation of the initial configuration 
- The last two arguments allow you to provide a starting spin configuration to continue a previous simulation

```python
import Kagome


J=1
L=5
Nx=L
Ny=L
diretcory='tutorial'
A=Kagome.Configuration(Nx,Ny, J,diretcory,seed_control=False,seed=1500,preload=False,oldconfig=None)
```

Then, we can start the simulation by looping over the temperatures and specify the number of Monte-Carlo steps and the steps for thermalization time (warmup procedure) to become stationnary, where we do not save the physical quantities (e.g. energy for heat capacity).

In the loop of temperatures we call the Monte-Carlo function with the argument is the number of MC steps  second is the temperature for the metropolis hastings acceptation rate then we specify if we want to record the energy every 1000 steps for the heat capacity measurement with the measure_capa boolean argument
Next if we want those steps to be hybrid with overrelaxation we specify overrelaxation=True and finally the overrelaxation_dose is the fraction of the lattice that will be 
rotated by an angle pi around the mean field
- we can then call the function saveconfig with the name of the file to be created with the informations

```python
Temp=[0.5,0.4,0.3]

decoherencetime=1000
numberMC=2000


for j,T in enumerate(Temp):
 
    A.Monte_Carlo(decoherencetime,T,measure_capa=False,overrelaxation=True,overrelaxation_dose=0.1)

    A.Monte_Carlo(numberMC,T,measure_capa=True,overrelaxation=True,overrelaxation_dose=0.1)


    name=f'simulationtempsetp{j}'
    
    A.saveconfig(name,T)
```
